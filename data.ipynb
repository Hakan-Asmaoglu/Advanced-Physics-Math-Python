```python
#Common libraries
#install first from command line, pip install numpy, pandas, matplotlib, 
#or install from the notebook, %pip install numpy, pandas, matplotlib,
#%pip install numpy
#%pip install pandas
#%pip install matplotlib
import numpy as np #Numerical calculations
import pandas as pd #Data manipulation  
import matplotlib.pyplot as plt #Graphics
```

```python
df = pd.read_csv('RawData.csv')
df.head()
df.info()
```

```python
#To access only a single column:
#df['Time (s)']

#First Panel
plt.figure(figsize=(15,8))
plt.subplot(3,1,1)
plt.plot(df['Time (s)'],df['Acceleration x (m/s^2)'])
plt.grid()
plt.title('TANYA')
plt.xlabel('Time')
plt.ylabel('a_x')
plt.axis([0, 10, -30, 30])

#Second Panel
plt.figure(figsize=(15,8))
plt.subplot(3,1,2)
plt.plot(df['Time (s)'],df['Acceleration y (m/s^2)'])
plt.grid()
plt.title('Acceleration y ')
plt.xlabel('Time')
plt.ylabel('a_y')
plt.axis([0, 10, -30, 30])

#Third Panel
plt.figure(figsize=(15,8))
plt.subplot(3,1,3)
plt.plot(df['Time (s)'],df['Acceleration z (m/s^2)'])
plt.grid()
plt.title('Acceleration z ')
plt.xlabel('Time')
plt.ylabel('a_z')
plt.axis([0, 10, -30, 30])
plt.show()
```
